```python
#Create plot area
fig, ax = plt.subplots()

#Set plot size
fig.set_size_inches(7, 5)

#Plot chart as above, but change the plot type from 'o' to '*' - givng us stars!
plt.plot(df['Frankreich(Euro/MWh)'],df['Niederlande(Euro/MWh)'],"o")

#add horizontal and vertical lines (crosshairs)
plt.plot([df['Frankreich(Euro/MWh)'].mean(),df['Frankreich(Euro/MWh)'].mean()],[90,20],'k-', linestyle = ":", lw=1)
plt.plot([20,90],[df['Niederlande(Euro/MWh)'].mean(),df['Niederlande(Euro/MWh)'].mean()],'k-', linestyle = ":", lw=1)

#Add labels to chart area
ax.set_title('Scatterplot Frankreich & Niederlande')
ax.set_xlabel("Frankreich")
ax.set_ylabel("Niederlande")

#Display the chart
plt.show()
```

```python
#Create our bar chart
plt.bar(df.Uhrzeit,height=df['Niederlande(Euro/MWh)'])
#Give it a title
plt.title("Niederlande(Euro/MWh)")
#rotate
plt.xticks(rotation=90)
#Give the x and y axes a title
plt.xlabel("Zeit")
plt.ylabel("Preis")
```

```python
#color
#Create an array of equal length to our bars
timeColors=['#034694','#001C58','#5CBFEB','#D00027',
              '#EF0107','#DA020E','#274488','#ED1A3B',
               '#000000','#091453','#60223B','#0053A0',
               '#E03A3E','#1B458F','#000000','#53162f',
               '#FBEE23','#EF6610','#C92520','#BA1F1A']
#Add a new argument, color, to our 'plt.bar()' method
plt.bar(df.Uhrzeit,height=df['Niederlande(Euro/MWh)'],color=timeColors)
#Give it a title
plt.title("Niederlande(Euro/MWh)")
#rotate
plt.xticks(rotation=90)
#Give the x and y axes a title
plt.xlabel("Zeit")
plt.ylabel("Preis")
```

```python
teamColours=['#f40206','#0560b5','#ce0000','#1125ff','#28cdff','#091ebc']
plt.pie(df['Niederlande(Euro/MWh)'],
        #Data labels are the team names in the dataFrame
       labels = df['Uhrzeit'],
        #Assign our colours list
       colors = teamColours,
        #Give a tidier angle to ur first data angle
        startangle = 90)
```

```python
#X – The data along the x axis
#Y – The data along the y axis
#Data – The dataframe we are reading from
#‘alpha’makes the dots see-through and shows how many values are on top of each other
sb.regplot(x="Frankreich(Euro/MWh)", y="Niederlande(Euro/MWh)",data=df, 
           color="purple",scatter_kws={'alpha':0.07})
```
